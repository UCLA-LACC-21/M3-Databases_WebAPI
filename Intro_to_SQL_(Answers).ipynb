```python
#Q14: show a list of all item names and their category number. If an item is in multiple categories, show the item once for each category number.
%%sql SELECT i.name as item, ic.cid as category_number 
FROM items i 
INNER JOIN item_category ic 
ON i.id = ic.pid;
```

```python
#Q15: show a list of all item names, their category name, and their subcategory name. If an item is in multiple categories, show the item once for each category.
%%sql SELECT i.name as item, c.main as category, c.sub as subcategory 
FROM items i 
INNER JOIN item_category ic 
ON i.id = ic.pid 
INNER JOIN categories c 
ON ic.cid = c.id;
```

```python
#Q16: For each category, show the number of subcategories
%%sql SELECT main as category, COUNT(*) as num_subcategories 
FROM categories 
GROUP BY category;
```

```python
#Q17: For each uid in orders, show the number of orders made by that user uid
%%sql SELECT user, COUNT(*) as num_orders 
FROM orders 
GROUP BY user;
```

```python
#Q18: for each item, show the number of times it has been ordered
%%sql SELECT i.name, SUM(oi.qty) as num_orders 
FROM items i 
INNER JOIN order_items oi 
ON i.id = oi.pid 
GROUP BY i.name;
```

```python
#Q19: what is the most popular item in the store?
%%sql SELECT i.name, SUM(oi.qty) as num_orders 
FROM items i 
INNER JOIN order_items oi 
ON i.id = oi.pid 
GROUP BY i.name 
ORDER BY num_orders DESC 
LIMIT 1;
```

```python
#Q20: show all items that have been ordered more than 50 times
%%sql SELECT i.name, SUM(oi.qty) as num_orders 
FROM items i 
INNER JOIN order_items oi 
ON i.id = oi.pid 
GROUP BY i.name 
HAVING num_orders > 50;
```

```python
#Q21: show a list of all items ordered in June 2021 that are currently out of stock
%%sql SELECT DISTINCT i.name
FROM items i 
INNER JOIN order_items oi 
ON i.id = oi.pid 
INNER JOIN orders o 
ON oi.oid = o.id 
WHERE o.time BETWEEN '2021-06-01' AND '2021-06-30' AND i.qty = 0;
```

```python
#Q22: show the number of orders placed every month starting from January 2021
%%sql SELECT MONTH(o.time) as month_number, COUNT(*) as number 
FROM order_items oi 
INNER JOIN orders o 
ON oi.oid = o.id 
WHERE o.time >= '2021-01-01' 
GROUP BY month_number;
```

```python
#Q23: how much money has been spent at the store?
%%sql SELECT SUM(i.price*oi.qty + o.tip)/100 as total 
FROM orders o 
INNER JOIN order_items oi 
ON o.id = oi.oid 
INNER JOIN items i 
ON oi.pid = i.id;
```

```python
#Q24: how much tip has been made?
%%sql SELECT SUM(tip)/100 as total_tip 
FROM orders;
```

```python
#Q25: who has the fastest delivered orders on average? what is their average delivery time?
%%sql SELECT user, AVG(delivered-time) as avg_deliver_time 
FROM orders 
GROUP BY user 
ORDER BY avg_deliver_time ASC 
LIMIT 1;
```

```python
#Q26: what is the average delivery time in each year-month? hopefully the delivery time is improving!
%%sql SELECT MONTH(time) as month_delivered, YEAR(time) as year_delivered, AVG(delivered-time) as avg_deliver_time 
FROM orders 
GROUP BY month_delivered, year_delivered 
ORDER BY avg_deliver_time DESC;
```

```python
#Q27: how many items have been ordered from each main category?
%%sql SELECT c.main as category, SUM(oi.qty) AS num_items_ordered 
FROM categories c 
INNER JOIN item_category ic 
ON c.id = ic.cid 
INNER JOIN items i 
ON ic.pid = i.id 
INNER JOIN order_items oi 
ON i.id = oi.pid 
GROUP BY main;
```

```python
#Q28: what are the top three most popular main categories?
%%sql SELECT c.main as category, SUM(oi.qty) AS num_items_ordered 
FROM categories c 
INNER JOIN item_category ic 
ON c.id = ic.cid 
INNER JOIN items i 
ON ic.pid = i.id 
INNER JOIN order_items oi 
ON i.id = oi.pid 
GROUP BY main 
ORDER BY num_items_ordered DESC 
LIMIT 3;
```

```python
#Q29: what are the top five most popular subcategories?
%%sql SELECT c.sub as subcategory, SUM(oi.qty) AS num_items_ordered 
FROM categories c 
INNER JOIN item_category ic 
ON c.id = ic.cid 
INNER JOIN items i 
ON ic.pid = i.id 
INNER JOIN order_items oi 
ON i.id = oi.pid 
GROUP BY subcategory 
ORDER BY num_items_ordered DESC 
LIMIT 5;
```

```python
#Q30: how much money has been spent on items from each subcategory?
%%sql SELECT c.sub as subcategory, SUM(oi.qty*i.price)/100 AS num_items_ordered 
FROM categories c 
INNER JOIN item_category ic 
ON c.id = ic.cid 
INNER JOIN items i 
ON ic.pid = i.id 
INNER JOIN order_items oi 
ON i.id = oi.pid  
GROUP BY subcategory;
```

```python
#Q31: which hour is the most popular for the store?
%%sql SELECT HOUR(o.time) as time_hour, AVG(oi.qty) AS num_items_ordered 
FROM orders o 
INNER JOIN order_items oi 
ON o.id = oi.oid 
GROUP BY time_hour 
ORDER BY num_items_ordered DESC 
LIMIT 1;
```

```python
#Q32: for each user, what was the first time they placed an order? show only the first time, once per user
%%sql SELECT user, MIN(time) as first_time 
FROM orders 
GROUP BY user;
```
